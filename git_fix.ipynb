```python
from pathlib import Path
import subprocess
import json

repo_root = Path('/workspaces/M2.-Despliegue-multientorno')
print('Repository root set to', repo_root)
```

```python
import subprocess
print(subprocess.run(['git', '-C', str(repo_root), 'branch', '--all'], capture_output=True, text=True).stdout)
print('---')
print(subprocess.run(['git', '-C', str(repo_root), 'remote', '-v'], capture_output=True, text=True).stdout)
print('---')
print(subprocess.run(['git', '-C', str(repo_root), 'status', '--short', '--branch'], capture_output=True, text=True).stdout)
```

```python
# Push branches to GitHub
for branch in ['develop', 'staging', 'main']:
    print(f'Pushing {branch}...')
    result = subprocess.run(['git', '-C', str(repo_root), 'push', '-u', 'origin', branch], capture_output=True, text=True)
    print(result.stdout)
    print(result.stderr)
    print('---')
```

```python
# Inspect git branch and status from the notebook environment
import subprocess

status = subprocess.run(['git', '-C', str(repo_root), 'status', '--short', '--branch'], capture_output=True, text=True)
print(status.stdout)
```

```python
# Create commits on develop, staging, and main with the correct config.js values.
configs = {
    'develop': {
        'environment': 'develop',
        'version': 'v0.1.0-dev',
        'updatedAt': '2026-04-09',
        'changes': [
            'Se creó la primera versión del laboratorio',
            'Se agregó una vista de entorno y versión',
            'Se preparó la estructura para trabajar con ramas'
        ]
    },
    'staging': {
        'environment': 'staging',
        'version': 'v0.1.0-staging',
        'updatedAt': '2026-04-10',
        'changes': [
            'Se agregó la rama staging para pruebas',
            'Se validó la página y configuración de entorno',
            'Preparación para publicar en GitHub Pages'
        ]
    },
    'main': {
        'environment': 'production',
        'version': 'v0.1.0',
        'updatedAt': '2026-04-10',
        'changes': [
            'Versión estable en main',
            'Configuración de GitHub Pages en staging',
            'Base para despliegue multientorno'
        ]
    }
}
config_path = repo_root / 'Laboratorio de Multientornos' / 'config.js'

for branch, cfg in configs.items():
    print('\nChecking out branch:', branch)
    result = subprocess.run(['git', '-C', str(repo_root), 'checkout', branch], capture_output=True, text=True)
    print(result.stdout, result.stderr)

    content = (
        'window.APP_CONFIG = {\n'
        f'  environment: "{cfg["environment"]}",\n'
        f'  version: "{cfg["version"]}",\n'
        f'  updatedAt: "{cfg["updatedAt"]}",\n'
        '  changes: [\n'
    )
    for change in cfg['changes']:
        content += f'    "{change}",\n'
    content += '  ]\n};\n'

    config_path.write_text(content)
    result = subprocess.run(['git', '-C', str(repo_root), 'add', str(config_path.relative_to(repo_root))], capture_output=True, text=True)
    print(result.stdout, result.stderr)
    result = subprocess.run(['git', '-C', str(repo_root), 'commit', '-m', f'Update config for {branch} environment'], capture_output=True, text=True)
    print(result.stdout, result.stderr)

subprocess.run(['git', '-C', str(repo_root), 'checkout', 'main'], capture_output=True, text=True)
print('\nFinished updating branches and returning to main.')
```

```python
# Inspect config.js contents in each branch to verify the environment values
for branch in ['develop', 'staging', 'main']:
    print('\nBranch:', branch)
    result = subprocess.run(['git', '-C', str(repo_root), 'show', f'{branch}:Laboratorio de Multientornos/config.js'], capture_output=True, text=True)
    print(result.stdout)
```
